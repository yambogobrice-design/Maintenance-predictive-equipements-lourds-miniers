# Finalisation du projet de maintenance prédictive

Ce notebook regroupe les étapes finales du projet : validation du code, génération du rapport, nettoyage des fichiers et exécution du pipeline complet.

## 1. Finaliser le code

Vérification, validation et sauvegarde de la version finale du code source. On s'assure que le code est propre, sans erreurs, et sauvegardé.

```python
# Exemple de validation du code Python
import subprocess
import sys
import glob

# Vérification de l'absence d'erreurs dans le dossier src
py_files = glob.glob('../src/*.py')
result = subprocess.run([sys.executable, '-m', 'py_compile'] + py_files, capture_output=True, text=True)
if result.returncode == 0:
    print('Code source validé : aucune erreur détectée.')
else:
    print('Erreurs détectées :')
    print(result.stderr)

# Sauvegarde du code (exemple : création d'une archive zip)
import shutil
shutil.make_archive('../code_final', 'zip', '../src')
print('Archive du code source créée.')
```

## 2. Générer un rapport

Utilisation de Python pour générer un rapport à partir des résultats du projet. On peut s'appuyer sur les notebooks précédents et le fichier rapport_M3.md.

```python
# Génération d'un rapport PDF à partir du markdown
import markdown
from fpdf import FPDF

# Charger le contenu du rapport markdown
with open('../rapport/rapport_M3.md', 'r') as f:
    md_content = f.read()

# Convertir le markdown en texte brut
html_content = markdown.markdown(md_content)

# Générer un PDF simple
pdf = FPDF()
pdf.add_page()
pdf.set_font('Arial', '', 12)
for line in html_content.split('\n'):
    pdf.multi_cell(0, 10, line)
pdf.output('../rapport/rapport_M3.pdf')
print("Rapport PDF généré dans le dossier rapport.")
```

## 3. Nettoyer les fichiers

Suppression des fichiers temporaires ou inutiles générés lors de l'exécution du projet.

```python
# Suppression des fichiers temporaires (exemple)
import os
import glob

temp_patterns = ["../src/*.pyc", "../src/__pycache__/*"]
for pattern in temp_patterns:
    for temp_file in glob.glob(pattern):
        try:
            os.remove(temp_file)
            print(f"Supprimé : {temp_file}")
        except Exception as e:
            print(f"Erreur suppression {temp_file} : {e}")
print("Nettoyage des fichiers temporaires terminé.")
```

## 4. Exécuter un pipeline

Définition et exécution du pipeline complet du projet à l'aide du script principal.

```python
# Exécution du pipeline principal du projet
import subprocess
import sys

result = subprocess.run([sys.executable, '../run_pipeline.py'], capture_output=True, text=True)
print(result.stdout)
if result.returncode != 0:
    print("Erreur lors de l'exécution du pipeline :")
    print(result.stderr)
else:
    print('Pipeline exécuté avec succès.')
```
